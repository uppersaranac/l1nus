```python
import pyarrow as pa
import pyarrow.csv as csv
import polars as pl
import numpy as np
```

```python
def read_tsv_to_table(file_path, column_names=None):
    # Read TSV into a PyArrow Table
    read_options = csv.ReadOptions(column_names=column_names)
    parse_options = csv.ParseOptions(delimiter='\t')
    table = csv.read_csv(file_path, read_options=read_options, parse_options=parse_options)
    return table

def write_table_to_arrowwriter(table, output_path):
    # Define features based on table schema
    with pa.OSFile(output_path, 'wb') as sink:
        with pa.RecordBatchStreamWriter(sink, table.schema) as writer:
            writer.write_table(table)

def pa_type_to_hf_type(pa_type):
    # Basic mapping from PyArrow types to HF dataset types
    if pa.types.is_integer(pa_type):
        return "int32"
    elif pa.types.is_floating(pa_type):
        return "float32"
    elif pa.types.is_string(pa_type):
        return "string"
    else:
        raise ValueError(f"Unsupported PyArrow type: {pa_type}")
```

```python
file_prefix = '/home/lyg/data/pubchem/arrow/cluster_100k'
# file_prefix = '/home/lyg/data/pubchem/arrow/cluster_6M'

table = read_tsv_to_table(f'{file_prefix}.txt', column_names=['cid','smiles','iupac','formula','num_atoms'])
# write_table_to_arrowwriter(table, 'cluster_100k.arrow')
```

```python
def partition_dataframe(df: pl.DataFrame, probs: tuple[float, float, float]) -> tuple[pl.DataFrame, pl.DataFrame, pl.DataFrame]:
    assert sum(probs) == 1.0, "Probabilities must sum to 1.0"
   
    # Generate random choices
    choices = np.random.choice([0, 1, 2], size=len(df), p=probs)
   
    # Create masks for each partition
    mask0 = pl.Series(choices == 0)
    mask1 = pl.Series(choices == 1)
    mask2 = pl.Series(choices == 2)

    df0 = df.filter(mask0)
    df1 = df.filter(mask1)
    df2 = df.filter(mask2)

    return df0, df1, df2
```

```python

# Convert PyArrow Table to Polars DataFrame
pl_df = pl.from_arrow(table)

# clean up the names
pl_df = pl_df.filter(pl_df['num_atoms'] >= 3)
pl_df = pl_df.filter(~pl_df['smiles'].str.contains(r'\.'))
pl_df = pl_df.filter(~pl_df['iupac'].str.contains(r'None'))
pl_df = pl_df.filter(~pl_df['smiles'].str.contains(r'\[2H\]'))

# Sort by 'num_atoms' and show the first 20 rows
sorted_df = pl_df.sort('num_atoms')
print(sorted_df.head(40))
# Randomly split the dataset
train_df, eval_df, test_df = partition_dataframe(sorted_df,[0.9, 0.05, 0.05])

# we need to convert to pyarrow tables as we have to cast columns of type string_view to 
# arrays of string. Polars only supports string_view and huggingface datasets only support string arrays.

# Convert Polars DataFrames to PyArrow Tables
train_table = train_df.to_arrow()
eval_table = eval_df.to_arrow()
test_table = test_df.to_arrow()

# Convert string_view columns to string columns
def convert_string_view_to_string(table):
    for column in table.schema:
        if pa.types.is_string_view(column.type):
            # Recreate the column as pa.string
            table = table.set_column(
                table.schema.get_field_index(column.name),
                column.name,
                pa.array(table[column.name].to_pylist(), type=pa.string())
            )
    return table

train_table = convert_string_view_to_string(train_table)
eval_table = convert_string_view_to_string(eval_table)
test_table = convert_string_view_to_string(test_table)

# Write each table to Arrow streaming format files
with pa.OSFile(f'{file_prefix}_train.arrow', 'wb') as sink:
    with pa.RecordBatchStreamWriter(sink, train_table.schema) as writer:
        writer.write_table(train_table)

with pa.OSFile(f'{file_prefix}_eval.arrow', 'wb') as sink:
    with pa.RecordBatchStreamWriter(sink, eval_table.schema) as writer:
        writer.write_table(eval_table)

with pa.OSFile(f'{file_prefix}_test.arrow', 'wb') as sink:
    with pa.RecordBatchStreamWriter(sink, test_table.schema) as writer:
        writer.write_table(test_table)
```
